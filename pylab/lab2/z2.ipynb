```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

db_data = load_diabetes()
db_features = db_data.feature_names

db_arr = db_data.data
X, y = db_arr[:, :-1], db_arr[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=221, shuffle=False)

from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(X_train, y_train)
y_pred = linReg.predict(X_test)
minval = min(y_test.min(), y_pred.min())
maxval = max(y_test.max(), y_pred.max())
fig1, fig2 = plt.subplots()
plt.scatter(y_test, y_pred)
plt.plot([minval, maxval], [minval, maxval], 'r')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Diabetes dataset')
plt.figure(0)

from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=221, shuffle=True)
linReg = LinearRegression()
linReg.fit(X_train, y_train)
y_pred = linReg.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f'MAE: {mae}, MSE: {mse}, MAPE: {mape}')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
plt.boxplot(y_train)
plt.ylabel('Target')
plt.figure(1)
plt.plot()
```
